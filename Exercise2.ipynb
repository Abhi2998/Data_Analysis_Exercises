```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.datasets import load_iris
```

```python
# Loading the iris dataset into dataframe
iris_data = load_iris(as_frame=True)
df = iris_data.frame
```

```python
#Displaying first 30 Values of the dataset
print(df.head(30)) 
```

```python
#Simple Random Sampling
sample_size=30
random_sampling=df.sample(sample_size,random_state=42)
print(random_sampling)
```

```python
sample_means=[] 

sample_size=30
for i in range(100):
    random_sampling=df.sample(sample_size,random_state=np.random.randint(1,199))
    sample_means.append(random_sampling['sepal length (cm)'].mean())
```

```python
# Plotting a histogram for the 100 sample means

plt.hist(sample_means, bins=20, color='red', edgecolor='black')
plt.title('Distribution of 100 Sample Means ')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.show()
```

```python
def systematic_sampling(df, percentage):
 
    step = int(len(df) / (len(df) * (percentage / 100)))
    return df.iloc[::step]
sampled_df = systematic_sampling(df, 20)
```

```python
plt.figure(figsize=(12, 6))

# Original dataset
plt.subplot(1, 2, 1)
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], alpha=0.7, label="Original")
plt.title('Original Dataset')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend()

# Sampled dataset
plt.subplot(1, 2, 2)
plt.scatter(sampled_df['sepal length (cm)'], sampled_df['sepal width (cm)'], color='orange', alpha=0.7, label="Sampled")
plt.title('Sampled Dataset (20%)')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend()

plt.tight_layout()
plt.show()
```
